```python
def summa(xs):
    s1 = 0
    s2 = 0
    s3 = 0
    s4 = 0
    for i in range(26):
        s1 += (xs[i]*w00[0][i])
        s2 += (xs[i]*w00[1][i])
        s3 += (xs[i]*w00[2][i])
        s4 += (xs[i]*w00[3][i])
    result = [s1,s2,s3,s4]
    print(result)
    return result
```

```python
def yc(s):
    result = []
    for i in s:
        if i > 0:
            result.append(1)
        else:
            result.append(-1)
    print(f"yc = {result}")
    return result
```

```python
def navcheno_or_not(n,y_e):
    k = 0
    for i in range(len(y00[n])):
        if (y00[n][i] == y_e[i]):
            k+=1
        else:
            k-=1
    if k == 4:
        print("Navcheno model")
    else:
        print("Ne navcheno model")
```
