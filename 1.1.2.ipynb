```python
import random
n = 6
m = 8
matrix = []
v1 = []
v2 = []
for i in range(m):
    time_array = []
    for j in range(n):
        time_array.append(random.uniform(-10,10))
    matrix.append(time_array)

r = [15,20,13,10,43,3,70,8]
    
for i in range(m):
    v2.append(i)
    v1.append(i)
for j in matrix:
    print(j)

print(v1)
print(v2)
print(r)
```

```python
array = []
for i in v1:
    for j in v2:
        array2 = []
        for k in range(n):
            array2.append((matrix[i][k]-matrix[j][k])**2)
        array.append(array2)
for i in array:
    print(i)
print(len(array))
```

```python
Evklid = []
for i in array:
    Evklid.append((sum(i))**(1/2))
print(Evklid)
print(len(Evklid))
```

```python
Evklid1 = []
for k in range(len(Evklid)):
    if (Evklid[k]!=0):
        Evklid1.append(Evklid[k])
print(Evklid1)
print(len(Evklid1))
```

```python
k = 1
res =[]
result = []
for i in range(len(Evklid1)):
    if (k<m-1):
        res.append(Evklid1[i])
        k+=1
    elif (k==m-1):
        res.append(Evklid1[i])
        result.append(res)
        res = []
        k=1
print(result)
print(len(result))
```

```python
Minim = []
for i in result:
    Minim.append(min(i))
print(Minim)
```

```python
vect1 = 0
vect2 = 0
Index = []
for i in result:
    for j in i:
        if Minim[vect1] == j:
            Index.append([vect1,vect2])
            print(Minim[vect1])
        vect2+=1
    vect2=0
    vect1+=1
print(Index)
```
