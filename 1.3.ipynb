```python
import numpy
a1 = [-1,0,4,-12,0.5]
a2 = [10,-3,0.5,2,3]
```

```python
def sum_dob(x1,x2):
    s_d = 0
    for i in range(len(x1)):
        s_d = s_d+x1[i]*x2[i]
    return s_d
```

```python
print(sum_dob(a1,a2))
```

```python
def mod(x1):
    mod1 = 0
    for i in range(len(x1)):
        mod1 += (x1[i]**2)
    mod1 = mod1**(1/2)
    return mod1
```

```python
print(mod(a1))
print(mod(a2))
```

```python
s = sum_dob(a1,a2)
dob = mod(a1)*mod(a2)
def cos(s1,dob1):
    c = s1/dob1
    return c
```

```python
print(cos(s,dob))
```

```python
def rad(x):
    arc = numpy.arccos(x)
    return arc
```

```python
radian = rad(cos(s,dob))
print(radian)
```

```python
def g(x):
    g1 = x*180/numpy.pi
    return g1
```

```python
print(g(radian))
```
